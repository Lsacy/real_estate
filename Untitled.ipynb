```python
all_data = []
for i in range(2,42):
    url = f"https://www.immonet.de/berlin/wohnung-mieten-seite{i}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    all_data.append([x.text for x in soup.find_all("div", "flex-grow-1 display-flex align-items-flex-end")])
```

```python
overall_data = []
for tags in all_data:
    for sub_tag in tags:
        overall_data.append(sub_tag)
print(overall_data)        
```

```python
first_refine = []
for strange_character in overall_data:
    first_refine.append(strange_character.replace("\n", "").replace("\t", "").replace("Miete zzgl. NK", "")
                       .replace("Wohnflächeca", "").replace("Miete inkl. NK", ""))
print(first_refine)    
```

```python
price = []
for elements in first_refine:
    price.append(elements.rpartition("m²")[0])
print(price)
```

```python
only_price = []
for num in price:
    only_price.append(num.rpartition("€")[0])
print(only_price)
```
